### Notebook to highlight the importance of economy of scale

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
# system A cost function
length_A = np.arange(1, 41, 3)
print(length_A)

cost_A = -1.05 * np.log(length_A) + 5.2
print(cost_A)
```

```python
# system B cost function
length_B = np.arange(1, 53, 3)
print(length_B)

cost_B = 30 / np.power(length_B, 0.95)
print(cost_B)
```

```python
# system A total cost
total_cost_A = length_A[-1] * cost_A[-1]
print(total_cost_A)

# system B total cost
total_cost_B = length_B[-1] * cost_B[-1]
print(total_cost_B)
```

```python
plt.plot(length_A, cost_A, linestyle='-', marker='o', label='cost of system A')
plt.plot(length_B, cost_B, linestyle='-', marker='o', label='cost of system B')
plt.title("Two connection systems cost variation")
plt.xlabel("Distance, line-mile")
plt.ylabel("Cost per line-mile")
plt.ylim(0, 8)
plt.yticks([0, 2, 4, 6, 8, 10])
plt.legend()

plt.show()
```

```python
# get the breakeven of the systems
length_delta = np.arange(1, 41, 0.2)
cost_A2 = -1.05 * np.log(length_delta) + 5.2
cost_B2 = 30 / np.power(length_delta, 0.95)
delta = cost_B2 - cost_A2

for d, c in zip(length_delta, delta):
    print([d, c])
```
